```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import os
%matplotlib inline
```

```python
data = pd.read_csv("C:/Users/amish/Desktop/Data Science/Advertising.csv")
data.head()
```


# Simple Linear Regression 

(Estimating sales wrt advertisement in TV)

```python
x = data['TV'].values.reshape(-1, 1)
y = data['Sales'].values.reshape(-1, 1)
```

```python
plt.figure(figsize=(16,8))
plt.scatter(x, y, c='black')
plt.xlabel("Money spent on TV ads ($)")
plt.ylabel("Sales ($)")
plt.show()
```

```python
# Splitting to Training and Testing dataset
from sklearn.model_selection import train_test_split

# 20% testing and 80% training dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
y_pred = reg.predict(x_test)
plt.figure(figsize=(16,8))
plt.scatter(x, y, c='black')
plt.plot(
    x_test,
    y_pred,
    c='blue',
    linewidth = 2
)
plt.xlabel("Money spent on TV ads ($)")
plt.ylabel("Sales ($)")
plt.show()
```

```python
reg.coef_ #calculating coefficients
```

```python
reg.intercept_ 
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
output = reg.predict([[230.1]])
output
```

# Multiple Linear Regression

```python
x = data.drop(['Sales'], axis = 1)
y = data['Sales'].values.reshape(-1, 1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
multiple_reg = LinearRegression()
multiple_reg.fit(x_train, y_train)
```

```python
y_pred = multiple_reg.predict(x_test)
```

```python
multiple_reg.coef_
```

```python
multiple_reg.intercept_
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
# taking input from the user 
print("Enter the amount you will invest on: ")
tv = float(input("TV : "))
radio = float(input("Radio : "))
newspaper = float(input("Newspaper : "))

# predicting the sales wrt all inputs
output = multiple_reg.predict([[tv],[radio],[newspaper]])
print("you will get Rs{:.2f} sales by advertising Rs{} on TV, Rs{} on Radio and Rs{} on newspaper."\
      .format(output[0][0] if output else "not predictable",[tv],[radio],[newspaper]))
```

```python
if not os.path.exists('models'):
    os.makedirs('models')
MODEL_PATH = "models/multiple_reg.sav"
pickle.dump(multiple_reg, open(MODEL_PATH, 'wb'))
```
